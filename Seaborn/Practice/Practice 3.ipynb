```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\SANJAY\\Downloads\\Seaborn\\dm_office_sales.csv")
```

```python
df.head()
```

## Countplot()

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.countplot(x='division', data=df)
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.countplot(data=df, x='level of education')
```

### Breakdown with another category with 'hue'

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.countplot(data=df, x='level of education', hue='training level')
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.countplot(data=df, x='level of education', hue='training level', palette='Set1')
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.countplot(data=df, x='level of education', hue='training level', palette='Paired')
```

## Barplot()

```python
plt.figure(figsize=(10, 6), dpi=200)
sns.barplot(x='level of education', y='salary', data=df, estimator=np.mean, ci='sd')
```

```python
plt.figure(figsize=(10, 6), dpi=200)
sns.barplot(x='level of education', y='salary', data=df, estimator=np.mean, ci='sd', hue='division')
```

```python
plt.figure(figsize=(10, 6), dpi=200)
sns.barplot(x='level of education', y='salary', data=df, estimator=np.mean, ci='sd', hue='division')
plt.legend(bbox_to_anchor=(1.05, 1))
```
