# Useful Methods

```python
df = pd.read_csv('tips.csv')
df.head()
```

```python
def last_four(num):
    return str(num)[-4:]
```

```python
df['CC Number'][0]
```

```python
last_four(df['CC Number'][0])
```

```python
df['last four'] = df['CC Number'].apply(last_four)
df.head()
```

```python
def yelp(num):
    if num < 10:
        return '$'
    elif num >= 10 and num < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['Doller Signs'] = df['total_bill'].apply(yelp)
df.head()
```

```python
df['total_bill'].mean()
```

```python
df['total_bill'].apply(lambda num: num*2).head()
```

```python
df.head()
```

```python
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return 'generous'
    else:
        return 'other'
```

```python
df['tip quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
df.head()
```

```python
df['tip quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
df.head()
```
