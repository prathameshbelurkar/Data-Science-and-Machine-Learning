```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0, 10)
y = x * 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label='x**2')
ax.plot(x, x**3, label='x**3')
ax.legend(loc=0)
# ax.legend(loc=1)
# ax.legend(loc=2)
# ax.legend(loc=3)
# ax.legend(loc=4)
# ax.legend(loc=(0.15, 0.8))

plt.show()
```

## Setting colors, linewidths, linetypes 

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, 'b.-')
ax.plot(x, x**3, 'g--')
```

## Suggested Approach: Use Keyword Arguments

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color='blue', alpha=0.5)
ax.plot(x, x+2, color="#8B008B")
ax.plot(x, x+3, color="#FF8C00")
```

## Line and Marker Styles 

## Linewidth

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(x, x-1, color='red', linewidth=0.25)
ax.plot(x, x-2, color='red', lw=0.5)
ax.plot(x, x-3, color='red', lw=1)
ax.plot(x, x-4, color='red', lw= 10)
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x-1, color="green", lw=3, linestyle='-') # solid
ax.plot(x, x-2, color="green", lw=3, ls='-.') # dash and dot
ax.plot(x, x-3, color="green", lw=3, ls=':') # dots
ax.plot(x, x-4, color="green", lw=3, ls='--') # dashes
```
