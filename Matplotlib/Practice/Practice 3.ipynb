## plt.subplots()

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('title')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(a, b)
axes[1].plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(a, b)
axes[1][1].plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(a, b)
axes[1][1].plot(x, y)

plt.tight_layout()
```

## Parameters on Subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

axes[0][0].plot(a, b)
axes[0][0].set_title('0 0 Title')

axes[1][1].plot(x, y)
axes[1][1].set_xlabel('1 1 Label')
axes[1][1].set_title('1 1 Title')

axes[0][1].plot(y, x)
axes[1][0].plot(b, a)

fig.suptitle('Figure level', fontsize=16)

plt.show()
```

### Manual Spacing on subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

axes[0][0].plot(a, b)
axes[1][1].plot(x, y)
axes[0][1].plot(b, a)
axes[1][0].plot(y, x)

fig.subplots_adjust(wspace=0.9, hspace=0.1)

plt.show()
```

### Exporting plt.subplots()

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

axes[0][0].plot(a, b)
axes[1][1].plot(x, y)
axes[0][1].plot(y, x)
axes[1][0].plot(b, a)

fig.savefig('subplots.png', bbox_inches='tight')

plt.show()
```
